```python
import numpy as np
import pandas as pd
```

```python
def missing_data(data):
    total = data.isnull().sum()
    percent = round(data.isnull().sum()/data.isnull().count()*100,2)
    tt = pd.concat((total, percent), axis=1, keys=['Total', 'Percent'])
    dtypes = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        dtypes.append(dtype)
    tt['Types'] = dtypes
    return np.transpose(tt)
```

```python
def most_frequent_values(data):
    count = data.count()
    tt = pd.DataFrame(count)
    tt.columns = ['Total']
    itms = []
    vals = []
    for col in data.columns:
        vc = data[col].value_counts()
        try:
            itms.append(vc.index[0])
            vals.append(vc.values[0])
        except Exception as ex:
            print(ex)
            itms.append(None)
            vals.append(None)
    tt['Most Frequent Item'] = itms
    tt['Frequency'] = vals
    tt['Frequency to Count %'] = round(vals/count*100,2)
    return np.transpose(tt)
```

```python
def unique_values(data):
    counts = data.count()
    tt = pd.DataFrame(counts)
    tt.columns = ['Total']
    uniq = []
    for col in data.columns:
        uniq.append(data[col].nunique())
    tt['Uniques'] = uniq
    return np.transpose(tt)
```
